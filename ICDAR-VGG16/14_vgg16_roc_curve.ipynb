```python
from sklearn.metrics import roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
```

```python
svm_roc_curve_save_path = '../../template/ICDAR-feature/icdar_vgg16_svm.npz'
```

```python
svm_roc_curve = np.load(svm_roc_curve_save_path)

svm_fpr = svm_roc_curve['fpr']
svm_tpr = svm_roc_curve['tpr']

svm_roc_auc_val = auc(svm_fpr, svm_tpr)
```

```python
# Plot ROC curve
plt.figure(figsize=(8, 8))
plt.plot(svm_fpr, svm_tpr, color='peru', lw=2, label=f'SVM Test AUC = {svm_roc_auc_val:.5f}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - SVM')
plt.legend()
plt.show()
```
