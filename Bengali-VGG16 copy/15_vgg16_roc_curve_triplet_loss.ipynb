```python
from sklearn.metrics import roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
```

```python
svm_roc_curve_save_path = '../../template/BENGALI-feature/bengali_vgg16_triplet_loss_svm_test.npz'
ann_roc_curve_save_path = '../../template/BENGALI-feature/bengali_vgg16_triplet_loss_ann_test.npz'
rf_roc_curve_save_path = '../../template/BENGALI-feature/bengali_vgg16_triplet_loss_rf_test.npz'
xgb_roc_curve_save_path = '../../template/BENGALI-feature/bengali_vgg16_triplet_loss_xgb_test.npz'
```

```python
svm_roc_curve = np.load(svm_roc_curve_save_path)
ann_roc_curve = np.load(ann_roc_curve_save_path)
rf_roc_curve = np.load(rf_roc_curve_save_path)
xgb_roc_curve = np.load(xgb_roc_curve_save_path)

svm_fpr = svm_roc_curve['fpr']
svm_tpr = svm_roc_curve['tpr']
ann_fpr = ann_roc_curve['fpr']
ann_tpr = ann_roc_curve['tpr']
rf_fpr = rf_roc_curve['fpr']
rf_tpr = rf_roc_curve['tpr']
xgb_fpr = xgb_roc_curve['fpr']
xgb_tpr = xgb_roc_curve['tpr']

svm_roc_auc_val = auc(svm_fpr, svm_tpr)
ann_roc_auc_val = auc(ann_fpr, ann_tpr)
rf_roc_auc_val = auc(rf_fpr, rf_tpr)
xgb_roc_auc_val = auc(xgb_fpr, xgb_tpr)
```

```python
# Plot ROC curve
plt.figure(figsize=(8, 8))
plt.plot(svm_fpr, svm_tpr, linestyle='-', color='red', lw=2, label=f'SVM Test AUC = {svm_roc_auc_val:.5f}')
plt.plot(ann_fpr, ann_tpr, linestyle='--', color='blue', lw=2, label=f'ANN Test AUC = {ann_roc_auc_val:.5f}')
plt.plot(rf_fpr, rf_tpr, linestyle=':', color='green', lw=2, label=f'RF Test AUC = {rf_roc_auc_val:.5f}')
plt.plot(xgb_fpr, xgb_tpr, linestyle='-.', color='peru', lw=2, label=f'XGB Test AUC = {xgb_roc_auc_val:.5f}')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Test - BENGALI - VGG16 - With Triplet Loss')
plt.legend()
plt.show()
```
