```python
import cv2 as cv
import os
```

```python
train_icdar_folder = "../template/ICDAR-data/train"
train_bengali_folder = "../template/BENGALI-data/train"
```

```python
signatures = [folder for folder in os.listdir(train_bengali_folder) if os.path.isdir(os.path.join(train_bengali_folder, folder))]
```

```python
signature_path = os.path.join(train_bengali_folder, signatures[0])
images = os.listdir(signature_path)
image_path = os.path.join(signature_path, images[0])
```

```python
print(image_path)
```

```python
img = cv.imread(image_path)
```

```python
resized_img = cv.resize(img, (64, 64))
```

```python
cv.imshow('Image', resized_img)
cv.waitKey(0)  # Wait for a key press before closing the window
cv.destroyAllWindows()  # Close all OpenCV windows
```

```python
output_path = "../template/image_bengali_64.png"
```
